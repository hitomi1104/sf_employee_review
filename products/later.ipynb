```python
h_location = []
h_description = []
h_price = []
for x in tqdm(range(0, 10)):
    url = 'https://www.rightmove.co.uk/property-for-sale/\
find.html?locationIdentifier=REGION%5E87490&index={}'.format(x)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    for house in soup.find_all('div', attrs={'class':'propertyCard-wrapper'}):
        try:
            h_location.append(house.find('address', attrs={'class':'propertyCard-address'}).text.strip())
        except:
            h_location.append(np.nan)

        try:
            h_description.append(house.find('div', attrs={'class':'propertyCard-description'}).text.strip())
        except:
            h_description.append(np.nan)

        try:
            h_price.append(house.find('div', attrs={'class':'propertyCard-priceValue'}).text)
        except:
            h_price.append(np.nan)

houses = pd.DataFrame({'Location': h_location,
                     'Description': h_description,
                     'Price': h_price})  
```

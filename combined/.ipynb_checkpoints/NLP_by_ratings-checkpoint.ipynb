```python
#Importing the basic librarires fot analysis

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")  #using style ggplot

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS
import re
```

```python
df = pd.read_csv("data/global_positive.csv")
df1 = pd.read_csv("data/global_negative.csv")
df2 = pd.read_csv("data/nlp.csv")
```

```python
df["pros_nlp"] = df2["pros_nlp"]
df1["cons_nlp"] = df2["cons_nlp"]
```

```python
df.drop(columns = ['Unnamed: 0'], inplace = True)
df.dropna(inplace = True)
df1.drop(columns = ['Unnamed: 0'], inplace = True)
df1.dropna(inplace = True)
```

```python
df['length'] = df['pros'].apply(len)
df1['length'] = df1['cons'].apply(len)
```

```python
df.head()
```

### Visualizations

```python
df['rating'] = df['rating'].apply(np.floor)
```

```python
df['rating'].unique()
```

```python
# how much percentage rating in the dataset

sns.countplot(df['rating'])
plt.show()
```

```python
# graph what the the relationship between Rating and Length
ax, fig = plt.subplots(nrows=1, ncols=1, figsize=(15, 5))
sns.lineplot(data=df,x="rating", y="length")
sns.lineplot(data=df1,x="rating", y="length")
# plt.legend()
plt.tight_layout()
```

```python
fig = px.scatter(df,x="rating",y="length", color="rating")
fig.write_html("images/rating_scatter.html")
```

```python
# Pros

df_5p=df[df["rating"]==5]
df_4p=df[df["rating"]==4]
df_3p=df[df["rating"]==3]
df_2p=df[df["rating"]==2]
df_1p=df[df["rating"]==1]
# Cons

df_5c=df1[df1["rating"]==5]
df_4c=df1[df1["rating"]==4]
df_3c=df1[df1["rating"]==3]
df_2c=df1[df1["rating"]==2]
df_1c=df1[df1["rating"]==1]
```
